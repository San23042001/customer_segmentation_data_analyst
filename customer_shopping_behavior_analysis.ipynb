```python
import pandas as pd
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
# Median review rating in each category is replaced to each missing value 

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# snake case the columns name
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ",'_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
## Feature Enginnering

# create a column age_group

labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].value_counts()
```

```python
## create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(5)
```

```python
df.info()
```

```python
df['purchase_frequency_days'] = (
    df['purchase_frequency_days']
    .round()
    .astype('int64')
)
```

```python
df['purchase_frequency_days'].isna().sum()
```

```python
import numpy as np
np.isinf(df['purchase_frequency_days']).sum()
```

```python
df['purchase_frequency_days'] = (
    df['purchase_frequency_days']
    .round()
    .astype('Int64')   # capital I
)
```
